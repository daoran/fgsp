## Spline Test

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
def interpolate_spline(xyz, n_points=20):
    tck, u = interpolate.splprep([xyz[:,0], xyz[:,1], xyz[:,2]], k=2, s=2)
    # x_knots, y_knots, z_knots = interpolate.splev(tck[0], tck)
    u_fine = np.linspace(0, 1, n_points)
    x_fine, y_fine, z_fine = interpolate.splev(u_fine, tck)
    return x_fine, y_fine, z_fine
```

```python
origin = np.array([0, 0, 1])
dest = np.array([10, 10, 1])
diff = (origin + dest) / 2

pivot = np.array([diff[0], diff[1], 5])

xyz = np.concatenate((origin, pivot, dest))
xyz = np.reshape(xyz, (3, -1))

x, y , z = interpolate_spline(xyz)

fig2 = plt.figure(2)
ax3d = fig2.add_subplot(111, projection='3d')
ax3d.plot(xyz[:,0], xyz[:,1], xyz[:,2], 'r*')
# ax3d.plot(xyz_spline[:,0], xyz_spline[:,1], xyz_spline[:,2], 'g')
ax3d.plot(x, y, z, 'g')
plt.show()
```
